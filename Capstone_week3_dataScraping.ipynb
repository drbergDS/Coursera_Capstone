```python
import pandas as pd

html_raw = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

pd.read_html will read in the tables on the page--after checking how many tables are in the list and checking their content, I determined the first  is the one we want to work with. I isolated this dataframe and removed "Not Assigned" boroughs. 

```python
len(html_raw)
toronto = html_raw[0]
display(toronto)
```

```python
# remove "not assigned" boroughs
print(len(toronto))
print(toronto["Borough"].value_counts())
toronto = toronto[toronto.Borough != "Not assigned"]
print(len(toronto))
```

Now that we have the "Not assigned" boroughs removed, we'll double-check that there are no repeated postal codes or neighborhoods with a "Not assigned" value

```python
display(toronto[toronto.duplicated(["Postal Code"], keep=False)])
display(toronto[toronto.Neighborhood=="Not Assigned"])
```
